```python
import matplotlib.pyplot as plt
import numpy as np
import os
%matplotlib inline
```

```python
A = np.load("tmp/data.npy")
```

```python
X1, X2 = A[0, :, :], A[1, :, :]
plt.scatter(X1[:, 0], X1[:, 1], c="r", alpha=0.5)
plt.scatter(X2[:, 0], X2[:, 1], c="b", alpha=0.5)
plt.show()
```

```python
f = np.linspace(-4, 5, 50)
xvals = os.listdir("tmp/xvals")
xvals.sort()

for xval in xvals:
    plt.figure()
    plt.scatter(X1[:, 0], X1[:, 1], c="r", alpha=0.5)
    plt.scatter(X2[:, 0], X2[:, 1], c="b", alpha=0.5)

    x = np.load("tmp/xvals/"+xval)
    for sub_x in x:
        a, b, c = sub_x
        g = -(a*f+c)/b
        plt.plot(f, g, "m-.", )

    # a, b, c = np.mean(x, axis=0)
    # g = -(a*f+c)/b
    # plt.plot(f, g, "k-")
    plt.title(xval)

    plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import os
%matplotlib inline
```

```python
A = np.load("tmp/data.npy")
```

```python
X1, X2 = A[0, :, :], A[1, :, :]
plt.scatter(X1[:, 0], X1[:, 1], c="r")
plt.scatter(X2[:, 0], X2[:, 1], c="b")
plt.show()
```

```python
f = np.linspace(-4, 5, 50)
counts = len(os.listdir("tmp/xvals"))

interval = 25
iters = list(range(counts)[::interval])
if iters[-1] != counts-1:
    iters.append(counts-1)

for i in iters:
    plt.figure()
    plt.scatter(X1[:, 0], X1[:, 1], c="r")
    plt.scatter(X2[:, 0], X2[:, 1], c="b")

    x = np.load("tmp/xvals/x{}.npy".format(i))
    for sub_x in x:
        a, b, c = sub_x
        g = -(a*f+c)/b
        plt.plot(f, g, "m-.", )

    a, b, c = np.mean(x, axis=0)
    g = -(a*f+c)/b
    plt.plot(f, g, "k-")
    plt.title("iterator {}".format(i+1))

    plt.show()
```
